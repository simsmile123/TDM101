## Question 2

```r
products_reviews_by_rating <- function(products_df, reviews_df, myrating) {
    merge_results <- merge(products_df, reviews_df, by="key")
    products_reviews_results <- merge_results[merge_results$rating >= myrating, ]
    return(products_reviews_results)
} 
```

## Question 3

the first step is creating the function and giving it a name, also it shows which parameters the function requires.
the second setp is to merge the products_df and reviews_df data frames based on a common key
the third step is to filter the merged data frame to include only reviews with a rating greater than or equal to myrating
finally, return the filtered data frame containing product reviews with the specified rating threshold

## Question 4

```r
my_selection <- products_reviews_by_rating(products, reviews, 4.5)
```
